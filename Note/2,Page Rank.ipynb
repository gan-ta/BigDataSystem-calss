```
#텍스트 단어 세기
text_file = sc.textFile("/FileStore/tables/test1.txt")
print(text_file.collect())
```

```
string = "this is a test"
print(string.split(" "))

#텍스트 파일 단어를 쪼갬
data1 = text_file.map(lambda line : line.split(" "))
print(data1.take(2))
```

```
#단어들만 모아놓음
data2 = text_file.flatMap(lambda line : line.split(" "))
print(data2.take(10))
```

```
#단어의 갯수를 파악하기 위한 준비과정
data3 = data2.map(lambda word:(word,1))
data3.take(10)
```

```
#단어의갯수를 도출
data4 = data3.reduceByKey(lambda a,b:a+b)
data4.take(10)
```

```
#단어의 갯수를 키로 설정
data5 = data4.map(lambda pair : (pair[1],pair[0]))
print(data5.take(10))
```

```
#키를 기준으로 정렬
data6 = data5.sortByKey(ascending = False) 
print(data6.collect())
```

```
#전체과정
text_file = sc.textFile("/FileStore/tables/test1.txt")

counts = text_file.flatMap(lambda line : line.split(" "))\
                  .map(lambda x : (x,1))\
                  .reduceByKey(lambda x,y : x + y)\
                  .map(lambda x : (x[1],x[0]))\
                  .sortByKey(ascending = False)
print(counts.collect())
counts.saveAsTextFile("/FileStore/tables/count.txt") #파일로 저장
```

```
#Page Rank
#중요 페이지와 덜 중요한 페이지 나눔
text_file = sc.textFile("/FileStore/tables/test1.txt")

print(text_file.count())
```

```
line_length = text_file.map(lambda s: len(s))
print(line_length.collect())
```

```
#reduce : 각 원소마다 더해버림 
total_length = line_length.reduce(lambda a,b : a+b) 
print(total_length)
```

```
data = sc.parallelize(['Carlo,5,3,3,4', 'Mokhtar,2,5,5,3', 'Jacques,4,2,4,5', 'Braden,5,3,2,5', 'Chris,5,4,5,1'])

data1 = data.map(lambda l : l.split(","))
print(data1.collect())
```

```
data2 = data1.map(lambda item: (item[0], int(item[1])+ int(item[2])+int(item[3])+int(item[4])))
print(data2.collect())
```

```
data3 = data2.map(lambda item : (item[0], item[1], item[1]/4))
print(data3.collect())
```

```
inputrdd = sc.parallelize([['maths',50], ['maths',60], ['english',65], ['english',85]])
print(inputrdd.collect())

#mapValues : values값의 조정
mapped = inputrdd.mapValues(lambda mark: (mark,1))
print(mapped.collect())
```

```
reduced = mapped.reduceByKey(lambda x,y : (x[0] + y[0], x[1] + y[1]))
print(reduced.collect())
```

```
average = reduced.map(lambda x : (x[0], x[1][0]/x[1][1]))
print(average.collect())
```

```
#PageRank 구현 시작
map_link = sc.parallelize([["MapR","Baidu"],["MapR", "Blogger"], ["Baidu","MapR"], ["Blogger", "Google"], ["Blogger","Baidu"], ["Google","MapR"]])
links = map_link.groupByKey()
print(links.collect())
print(list((k,list(v)) for (k,v) in links.collect()))
```

```
ranks = links.map(lambda pairs:(pairs[0] ,1))
print(ranks.collect())
```

```
#join을 사용하여 키에 대한 value를 이어붙힘
cvalues = links.join(ranks)
print(cvalues.collect())
print(list((k,list(v)) for (k,v) in cvalues.collect()))
print(list((k,list(v[0]),v[1]) for (k,v) in cvalues.collect()))
```

```
#urls는 화살표를 받은 urls, rank는 시작점에서의 초기 rank
def computeContribs(urls, rank):
  num_urls = len(urls)
  for url in urls:
    yield (url, rank/num_urls)
```

```
a = computeContribs([1,2,3],1)
for i in a:
  print(i)
```

```
contribs = links.join(ranks).flatMap(lambda url_urls_rank: computeContribs(url_urls_rank[1][0], url_urls_rank[1][1]))
print(contribs.collect())
```

```
new_rank = contribs.reduceByKey(lambda x,y : x+y).collect()
print(new_rank)
#기존에 있던 rank에서 식을 대입하여 갱신
new_rank1 = contribs.reduceByKey(lambda x,y : x+y).mapValues(lambda rank: 0.15+0.85 * rank).collect()
print(new_rank1)
```
