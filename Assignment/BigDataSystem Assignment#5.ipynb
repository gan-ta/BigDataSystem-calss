```python
import pandas as pd
import numpy as np
movies = pd.read_csv("http://bit.ly/imdbratings")
movies.head()
```

```python
movies_content_rating = movies[['content_rating']].drop_duplicates().reset_index(drop=True)
movies_content_rating
```

```python
movies_genre = movies[['genre']].drop_duplicates().reset_index(drop=True)
movies_genre
```

```python
movies_actors_list = movies[['actors_list']].drop_duplicates().reset_index(drop=True)
movies_actors_list
```
